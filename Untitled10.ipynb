```python
import open3d as o3d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the PLY file
ply_file = "AT_PCD_20240704_063525.ply"
pcd = o3d.io.read_point_cloud(ply_file)

# Step 2: Extract and save data to CSV
pcd_np = np.asarray(pcd.points)
csv_file = "point_cloud_pcd18.csv"
pd.DataFrame(pcd_np, columns=['x', 'y', 'z']).to_csv(csv_file, index=False)

# Step 3: Load CSV for plotting
data = pd.read_csv(csv_file)



# Define x and y-axis ROI
x_min = -72
x_max = -4.5
y_min = 0.5
y_max = 110

# Step 4: Define multiple y-values and tolerance 
y_values = [int(y_min+(y_max-y_min)*0.1), int(y_min+(y_max-y_min)*0.2), int(y_min+(y_max-y_min)*0.3),int(y_min+(y_max-y_min)*0.4),int(y_min+(y_max-y_min)*0.5),int(y_min+(y_max-y_min)*0.6), int(y_min+(y_max-y_min)*0.7),int(y_min+(y_max-y_min)*0.8),int(y_min+(y_max-y_min)*0.9)] 
tolerance = 0.028

# Filter data based on x and y-axis range
filtered_data = data[(data['x'] >= x_min) & (data['x'] <= x_max) & (data['y'] >= y_min) & (data['y'] <= y_max)]

# Crop the point cloud based on x and y-axis range
pcd_filtered = pcd.select_by_index(
    np.where((pcd_np[:, 0] >= x_min) & (pcd_np[:, 0] <= x_max) & (pcd_np[:, 1] >= y_min) & (pcd_np[:, 1] <= y_max))[0]
)

# Initialize Open3D LineSet components
line_points = []
lines = []

# Function to identify continuous segments in Z-axis values
def get_continuous_segments(points, axis_index, tolerance):
    segments = []
    if(len(points)==0):
        return segments
    segment = [points[0]]
    for i in range(1, len(points)):
        if abs(points[i, axis_index] - points[i - 1, axis_index]) <= tolerance:
            segment.append(points[i])
        else:
            if len(segment) > 1:
                segments.append(segment)
            segment = [points[i]]
    if len(segment) > 1:
        segments.append(segment)
    return segments

# Function to process each y-value
def process_y_value(y_value, tolerance, min_segment_length):
    filtered_data_y = filtered_data[(filtered_data['y'] >= y_value - tolerance) & (filtered_data['y'] <= y_value + tolerance)]
    filtered_np = filtered_data_y[['x', 'y', 'z']].to_numpy()
    tolerance_gap = 0.25
    segments = get_continuous_segments(filtered_np, 2, tolerance_gap)

    considered_segments = []
    if(len(segments)==0):
        return [0] 
    # Filter segments based on min_segment_length
    for segment in segments:
        if len(segment) >= min_segment_length:
            considered_segments.append(segment)

    # Plot only the considered lines and gaps
    plt.figure(figsize=(10, 6))

    # Plot the considered lines
    for idx, segment in enumerate(considered_segments):
        if len(segment) >= 2:
            segment_np = np.array(segment)
            plt.plot(segment_np[:, 0], segment_np[:, 2], marker='o', label=f'Line {idx + 1}')

    # Find and plot gaps between considered segments
    gap1_distances = []
    for i in range(len(considered_segments) - 1):
        distance = abs(considered_segments[i + 1][0][0] - considered_segments[i][-1][0])
        print(f"Distance between line {i + 1} and line {i + 2}: {distance}")
        if i == 0:  # Record only the first gap (gap1)
            gap1_distances.append(distance)
        # Annotate the plot with line number
        mid_x = (considered_segments[i][-1][0] + considered_segments[i + 1][0][0]) / 2
        mid_z = (considered_segments[i][-1][2] + considered_segments[i + 1][0][2]) / 2
        plt.annotate(f"Gap {i + 1}", (mid_x, mid_z), textcoords="offset points", xytext=(0, 10), ha='center')

    for segment in considered_segments:
        if len(segment) >= 2:
            for i in range(len(segment) - 1):
                lines.append([len(line_points), len(line_points) + 1])
                line_points.append(segment[i])
            line_points.append(segment[-1])
        else:
            print("Segment is too short and is not plotted.")

    plt.xlabel('X-axis')
    plt.ylabel('Z-axis')
    plt.title(f'Plot of Z for y ≈ {y_value} ± {tolerance}')
    plt.legend()
    plt.show()

    return gap1_distances

# Collect gap1 distances for all y-values
all_gap1_distances = []
for y_value in y_values:
    gap1_distances = process_y_value(y_value, tolerance, min_segment_length=10.0)  # Adjust the minimum segment length as needed
    all_gap1_distances.extend(gap1_distances)

# Calculate and print the average of gap1 distances
if all_gap1_distances:
    average_gap1 = np.mean(all_gap1_distances)
    print(f"Average of gap1 distances: {average_gap1}")
else:
    print("No gap1 distances found.")

# Create a LineSet from the filtered points
line_set = o3d.geometry.LineSet()
line_set.points = o3d.utility.Vector3dVector(line_points)
line_set.lines = o3d.utility.Vector2iVector(lines)
colors = [[1, 0, 0] for _ in range(len(lines))]
line_set.colors = o3d.utility.Vector3dVector(colors)

# Visualize the original point cloud and the line set
o3d.visualization.draw_geometries([pcd_filtered, line_set], window_name="Point Cloud with Filtered Lines")
```
